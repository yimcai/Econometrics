### Q2


(b) To check that the series εxt and εyt are uncorrelated, regress εyt on a constant and εxt. Report the t-value and p-value of the slope coefficient.

```python
epsi = df[["X", "Y" ]].diff().dropna()
epsi = epsi.rename(columns={"X":"epsi_x", "Y":"epsi_y"})
```

```python
y, X = dmatrices("epsi_y ~ epsi_x", data= epsi)
```

```python
mod1 = sm.OLS(y, X).fit()
```

```python
print (mod1.summary())
```


(c) Extend the analysis of part (b) by regressing εyt on a constant, εxt, and three lagged values of εyt and of εxt. Perform the F-test for the joint insignificance of the seven parameters of εxt and the three lags of εxt and εyt. Report the degrees of freedom of the F-test and the numerical outcome of this test, and draw your conclusion. Note: The relevant 5% critical value is 2.0.

```python
epsi.head(2)
```

```python
#lagged 1
epsi_lagged1 = epsi.shift(1)
epsi_lagged1 = epsi_lagged1.rename(columns= {"epsi_x": "epsi_x_lagged1" ,
                                             "epsi_y": "epsi_y_lagged1"})
#lagged 2
epsi_lagged2 = epsi.shift(2)
epsi_lagged2 = epsi_lagged2.rename(columns= {"epsi_x": "epsi_x_lagged2" ,
                                             "epsi_y": "epsi_y_lagged2"})
#lagged 3
epsi_lagged3 = epsi.shift(3)
epsi_lagged3 = epsi_lagged3.rename(columns= {"epsi_x": "epsi_x_lagged3" ,
                                             "epsi_y": "epsi_y_lagged3"})
```

```python
epsi_lagged = pd.concat([epsi, epsi_lagged1, epsi_lagged2, epsi_lagged3], axis= 1)
```

```python
y2, X2 = dmatrices("epsi_y ~ epsi_x + epsi_x_lagged1 + epsi_x_lagged2 + epsi_x_lagged3 + epsi_y_lagged1 + epsi_y_lagged2 + epsi_y_lagged3",
          data = epsi_lagged)
```

```python
mod2 = sm.OLS(y2, X2).fit()
```

```python
print (mod2.summary())
```

### Regress on constant model and $\epsilon_{xt}$

```python
y3, X3 = dmatrices("epsi_y ~ 1 + epsi_x",
          data = epsi_lagged)
```

```python
mod3 = sm.OLS(y3, X3).fit()
print (mod3.summary())
```
