```python
import pandas as pd 
import numpy as np 
import statsmodels.api as sm
from patsy import  dmatrices
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
df = pd.read_excel("Test6_data.xlsx")
```

**(a) Make time series plots of the CPI of the Euro area and the USA, and also of their logarithm log(CPI) and of the two monthly inflation series DP = âˆ†log(CPI). What conclusions do you draw from these plots?**


```python
fig, ax = plt.subplots(3, 1, figsize = (9, 16))
df_plot = df.set_index("TREND").copy()
df_plot[["CPI_EUR", "CPI_USA"]].plot(ax = ax[0])
df_plot[["LOGPEUR", "LOGPUSA"]].plot(ax = ax[1])
df_plot[["DPEUR", "DPUSA"]].plot(ax = ax[2])

ax[0].set_title("CPI of Europe and USA")
ax[1].set_title("Logarithm of CPI of Europe and USA")
ax[2].set_title("Monthly Inflation of Europe and USA")
```

>Conclusion:

>1. There exists a deterministic trend for CPI in both Europe and USA, so as their logarithm terms. 

>2. The CPI for both Europe and USA are non-stationary as their means varies with time. 

>3. The inflation rate, however, seems stationary and thus can be used for time-series analysis. 

>4. The inflation rate of Europe and USA seem correlated.

**(b) Perform the Augmented Dickey-Fuller (ADF) test for the two log(CPI) series. In the ADF test equation, include a constant (Î±), a deterministic trend term (Î²t), three lags of DP = âˆ†log(CPI) and, of course, the variable of interest log(CPItâˆ’1). Report the coefficient of log(CPItâˆ’1) and its standard error and t-value, and draw your conclusion.**

```python
df_lagged1_term = df[["LOGPEUR", "LOGPUSA", "DPEUR", "DPUSA"]].shift(1).rename(columns = {"LOGPEUR": "LOGPEUR_L1", 
                                                                                          "LOGPUSA": "LOGPUSA_L1",
                                                                                          "DPEUR" : "DPEUR_L1", 
                                                                                          "DPUSA": "DPUSA_L1"})

df_lagged2_term = df[["DPEUR", "DPUSA"]].shift(2).rename(columns= {"DPEUR":"DPEUR_L2", 
                                                                   "DPUSA":"DPUSA_L2"})

df_lagged3_term = df[["DPEUR", "DPUSA"]].shift(3).rename(columns = {"DPEUR":"DPEUR_L3",
                                                                    "DPUSA":"DPUSA_L3"})
df_lagged = pd.concat([df, df_lagged1_term, df_lagged2_term, df_lagged3_term], axis =1)
```

```python
y_eur, X_eur = dmatrices("DPEUR ~ TREND+LOGPEUR_L1 +DPEUR_L1+DPEUR_L2+DPEUR_L3", df_lagged)
y_usa, X_usa = dmatrices("DPUSA ~TREND+LOGPUSA_L1+DPUSA_L1+DPUSA_L2+DPUSA_L3" , df_lagged)
```

```python
eur_mod = sm.OLS(y_eur, X_eur).fit()
usa_mod = sm.OLS(y_usa, X_usa).fit()
```

```python
print (eur_mod.summary())
```

```python
print (usa_mod.summary())
```

>As the results indicate, for both variables, the ADF statistic is greater than the critical value of âˆ’3.5.
Therefore, the non-stationarity hypothesis is not rejected. 

**(c) As the two series of log(CPI) are not cointegrated (you need not check this), we continue by modelling the monthly inflation series DPEUR = âˆ†log(CPIEUR) for the Euro area. Determine the sample autocorrelations and the sample partial autocorrelations of this series to motivate the use of the following AR model: DPEURt = Î± + Î²1DPEURtâˆ’6 + Î²2DPEURtâˆ’12 + Îµt . Estimate the parameters of this model (sample Jan 2000 - Dec 2010).**

```python
dpeur_l6 = df[["DPEUR"]].shift(6).rename(columns = {"DPEUR": "DPEUR_L6"})
dpeur_l12 = df[["DPEUR"]].shift(12).rename(columns = {"DPEUR": "DPEUR_L12"})
df_lagged_c = pd.concat([df_lagged, dpeur_l6, dpeur_l12], axis= 1)
```

```python
from statsmodels.tsa.stattools import acf, pacf
```

```python
dpeur = df_lagged_c[df_lagged_c.TREND <= 132].DPEUR.dropna().values
```

```python
acf(dpeur, nlags= 12)[1:]
```

```python
pacf(dpeur, nlags= 12)[1:]
```

>The lags with largest ACF and PACF were found at lag6 and lag12

>Estimation:

```python
y_c, X_c = dmatrices("DPEUR ~ DPEUR_L6+DPEUR_L12", data= df_lagged_c[df_lagged_c.TREND <= 132])
```

```python
mod_c = sm.OLS(y_c, X_c).fit()
print (mod_c.summary())
```

**(d) Extend the AR model of part (c) by adding lagged values of monthly inflation in the USA at lags 1, 6, and 12. Check that the coefficient at lag 6 is not significant, and estimate the ADL model DPEURt = Î± + Î²1DPEURtâˆ’6 + Î²2DPEURtâˆ’12 + Î³1DPUSAtâˆ’1 + Î³2DPUSAtâˆ’12 + Îµt (sample Jan 2000 - Dec 2010).**

```python
dpusa_l6 = df[["DPUSA"]].shift(6).rename(columns = {"DPUSA": "DPUSA_L6"})
dpusa_l12 = df[["DPUSA"]].shift(12).rename(columns = {"DPUSA": "DPUSA_L12"})
df_lagged_d = pd.concat([df_lagged_c, dpusa_l6, dpusa_l12], axis= 1)
```

```python
y_d, X_d = dmatrices("DPEUR~DPEUR_L6+DPEUR_L12+DPUSA_L1+DPUSA_L6+DPUSA_L12", df_lagged_d[df_lagged_c.TREND <= 132])
```

```python
mod_d = sm.OLS(y_d, X_d).fit()
print (mod_d.summary())
```

>The p-value for DPUSA at lag 6 ("DPUSA_L6") is 0.308, which is not significant at 95% confidence level. Therefore we keep variable "DPUSA_L6" out of the model, and new estimation process goes as follows:

```python
y_d2, X_d2 = dmatrices("DPEUR~DPEUR_L6+DPEUR_L12+DPUSA_L1+DPUSA_L12", df_lagged_d[df_lagged_c.TREND <= 132])
mod_d2 = sm.OLS(y_d2, X_d2).fit()
print (mod_d2.summary())
```

**
(e) Use the models of parts (c) and (d) to make two series of 12 monthly inflation forecasts for 2011. At each month, you should use the data that are then available, for example, to forecast inflation for September 2011 you can use the data up to and including August 2011. However, do not re-estimate the model and use the coefficients as obtained in parts (c) and (d). For each of the two forecast series, compute the values of the root mean squared error (RMSE), mean absolute error (MAE), and the sum of the forecast errors (SUM). Finally, give your interpretation of the outcomes.**

```python
def predict_future(trend, mod = "C", data = df_lagged_d):
    if mod == "C":
        mod = mod_c
        exogs = ["const","DPEUR_L6", "DPEUR_L12"]
    elif mod == "D":
        mod = mod_d2
        exogs = ["const","DPEUR_L6", "DPEUR_L12", "DPUSA_L1", "DPUSA_L12"]
    else:
        raise Exception("Model does not exist")
    data = sm.add_constant(data)
    predicted_values = mod.predict(data[data.TREND == trend][exogs])
    return predicted_values.values[0]
```

```python
#predicted values 
trends = range(133, 145)
mod_c_predicted = []
mod_d_predicted =[]
for trend in trends:
    mod_c_predicted.append(predict_future(trend, mod = "C"))
    mod_d_predicted.append(predict_future(trend, mod= "D"))
#actual values 
real_dpeur = df_lagged_d[df_lagged_d.TREND.isin(trends)].DPEUR.values
```

```python
fig, ax = plt.subplots(1, 2, figsize = (16, 6))
xs = range(1, 13)
ax[0].plot(xs, mod_c_predicted, label = "model C predicted", linestyle = "--", marker = "." ,c = "blue")
ax[0].plot(xs, mod_d_predicted, label = "model D predicted", marker ="o", c= "orange")
ax[0].plot(xs, real_dpeur, label = "Actual DPEUR", c = "red")
ax[0].legend()
ax[0].set_title("2011 Europe Monthly Inflation Forecast")
ax[0].set_xlabel("Month")

ax[1].plot(xs, real_dpeur- mod_c_predicted , label = "model C residuals", linestyle = "--", c = "blue")
ax[1].plot(xs, real_dpeur-  mod_d_predicted, label = "model D residuals", marker ="o", c= "orange")
ax[1].set_title("2011 Europe Monthly Inflation Forecast residuals")
ax[1].set_xlabel("Month")
ax[1].legend()
```

```python
RMSE_C  = np.sqrt(sum([resid**2 for resid in (real_dpeur- mod_c_predicted)])/12)
RMSE_D  = np.sqrt(sum([resid**2 for resid in (real_dpeur- mod_d_predicted)])/12)
MAE_C = sum([np.abs(resid) for resid in (real_dpeur- mod_c_predicted)])/12
MAE_D = sum([np.abs(resid) for resid in (real_dpeur- mod_d_predicted)])/12
SUM_C = sum([resid for resid in (real_dpeur- mod_c_predicted)])
SUM_D = sum([resid for resid in (real_dpeur- mod_d_predicted)])
```
